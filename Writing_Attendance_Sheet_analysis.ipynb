```python
import pandas as pd; import matplotlib.pyplot as plt
#week11 = pd.read_excel("A:\Minerva KGI\Academic_Internship\Writing_analysis\Writing_SSS_Attendance_Tracker.xlsx", sheet_name=1)
```

```python
#absences_list = data.drop(columns=["Email", "Date of makeup work submission (only required for students who did not show up at all)", "Type of absence", "Notes ", "Date of absence (double click to select from calendar)"])
absences = []
for i in range(1,11):
    data = pd.read_excel("A:\Minerva KGI\Academic_Internship\Writing_analysis\Writing_SSS_Attendance_Tracker.xlsx", sheet_name=i)
    absent=list(data.loc[:,'Present for Session?'])
    counter=0
    for x in absent:
        if x == "No":
            counter+=1
    absences.append(counter)
    
print(absences)
```

```python
test_absences = absences[:]
test_absences.reverse()
weeks_list = range(2,12,1)
absences_ratios = []
for i in range(len(test_absences)):
    absences_ratios.append(test_absences[i]/34)
    
    
print(len(weeks_list), len(absences))
print(absences_ratios)
```

```python
plt.plot(weeks_list, test_absences, marker='o')
plt.ylabel("Number of missing students"); plt.xlabel("Week")
plt.grid(alpha=.2)
```

```python
plt.plot(weeks_list, absences_ratios, marker='o')
plt.ylabel("Ratio of missing students"); plt.xlabel("Week")
plt.ylim(0, 1)
plt.grid(alpha=.2)
```

```python
engagement = []
for i in range(1,11):
    data = pd.read_excel("A:\Minerva KGI\Academic_Internship\Writing_analysis\Writing_SSS_Attendance_Tracker.xlsx", sheet_name=i)
    engaged=data.loc[:,'Engaged?']
    engaged = engaged.fillna("No")
    engaged = list(engaged)
    counter=0
    for x in engaged:
        if x == "No":
            counter+=1
    engagement.append(counter)
```

```python
test_engagement = engagement[:]
test_engagement.reverse()
print(test_engagement)

    
```

```python
plt.plot(weeks_list, test_engagement, marker='o')
plt.ylabel("Number of unengaged students"); plt.xlabel("Week")
plt.grid(alpha=.2)
```
